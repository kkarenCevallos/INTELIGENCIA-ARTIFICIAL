```python
import pandas as pd
url="/content/daily_weather.csv"
df=pd.read_csv(url)
df.head()
```

```python
df.describe()
```

```python
df['relative_humidity_3pm'].value_counts() > 36
```

```python
# Crear la nueva columna con la condición especificada
df['Prediction_Weather'] = df['relative_humidity_3pm'].apply(lambda x: 0 if x > 36 else 1)
print(df)
```

```python
df['Prediction_Weather'].value_counts()
```

```python
df = df.dropna(subset=['air_pressure_9am'])
df = df.dropna(subset=['air_temp_9am'])
df = df.dropna(subset=['avg_wind_direction_9am'])
df = df.dropna(subset=['avg_wind_speed_9am'])
df = df.dropna(subset=['max_wind_direction_9am'])
df = df.dropna(subset=['max_wind_speed_9am'])
df = df.dropna(subset=['rain_accumulation_9am'])
df = df.dropna(subset=['rain_duration_9am'])
```

```python
df.shape
```

```python
tmp=list()
for apro in df["relative_humidity_3pm"]:
    if apro < 36:
        tmp.append(0)
    else:
        tmp.append(1)

df["target"] = tmp
df.head()
```

```python
x = df.drop(["relative_humidity_3pm","Prediction_Weather","target","number"], axis=1)
x.head()
```

```python
y=df["target"]
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
yPrediccion=model.predict(x_test)
```

```python
yPrediccion
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,yPrediccion)
```
